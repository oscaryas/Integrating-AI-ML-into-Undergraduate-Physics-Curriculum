## Eigenvalues and Eigenvectors

```python
#Matrix from example in lecture
M = np.array([[1,0,3],[0,-2,0],[3,0,1]]) 
print(M)
```

```python
#Compute the eigenvalues of matrix above, and print the values 
w, v = np.linalg.eig(M)
```

```python
#Print Eigenvalues
print(w)
```

```python
#Print Eigenvectors
print(v)
```

```python
#invert the matrix of eigenvectors and print it
C = v
C_inv = np.linalg.inv(C)
print(C_inv)
```

```python
#Diagonalization of the matrix with the eigenvalues/eigenvectors
D = np.diag(w)
print(D)
```

```python
#Perform proper multiplying of matrices
np.dot(C_inv,np.dot(M,C))
```

## Creating Multi-Dimensional Tensors

```python
# Import TensorFlow and other necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Check TensorFlow version
print(f'TensorFlow Version: {tf.__version__}')
```

```python
# Define a scalar
scalar = 5

# Define a vector
vector = np.array([1, 2, 3])

# Define a matrix
matrix = np.array([[1, 2], [3, 4], [5, 6]])

# Define a 3rd-order tensor
tensor_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Tensor Addition
tensor_a = np.array([[[1, 1], [1, 1]], [[1, 1], [1, 1]]])
tensor_b = np.array([[[2, 2], [2, 2]], [[2, 2], [2, 2]]])
tensor_sum = tensor_a + tensor_b

# Scalar Multiplication
scalar_mult = 3 * tensor_a

# Tensor Product
vector1 = np.array([1, 2])
vector2 = np.array([3, 4])
tensor_product = np.outer(vector1, vector2)  # Equivalent to the tensor product of two vectors

# Print results
print("Tensor Addition:\n", tensor_sum)
print("Scalar Multiplication:\n", scalar_mult)
print("Tensor Product:\n", tensor_product)
```

In TensorFlow, tensors are multi-dimensional arrays. Below we will create a few examples of 2D and 3D tensors.

```python
# 2D tensor (matrix)
tensor_2d = tf.constant([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print("2D Tensor (Matrix):\n", tensor_2d)

# 3D tensor (e.g., a batch of 2D images)
tensor_3d = tf.random.normal([3, 3, 3])  # Create a 3x3x3 tensor with random values
print("3D Tensor:\n", tensor_3d)
```

## Visualizing Tensors

```python
# Import necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a helper function to visualize 3D tensor slices
def plot_tensor_slices(tensor, axis=0):
    """
    Function to visualize slices of a 3D tensor along a specified axis.

    Args:
    tensor (np.ndarray): A 3D tensor to visualize.
    axis (int): Axis along which to slice the tensor (0, 1, or 2).
    """
    # Check if the tensor is 3D
    assert tensor.ndim == 3, "The input tensor must be 3-D"

    # Number of slices depends on the axis chosen
    num_slices = tensor.shape[axis]
    
    fig, axes = plt.subplots(1, num_slices, figsize=(15, 5))
    for i in range(num_slices):
        if axis == 0:
            slice_ = tensor[i, :, :]
        elif axis == 1:
            slice_ = tensor[:, i, :]
        else:
            slice_ = tensor[:, :, i]

        ax = axes[i] if num_slices > 1 else axes
        ax.imshow(slice_, cmap='viridis')
        ax.set_title(f'Slice {i} along axis {axis}')
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# Define a 3D tensor using TensorFlow
tensor_3d_tf = tf.random.uniform(shape=(4, 4, 4), minval=0, maxval=1)

# Convert TensorFlow tensor to NumPy array for visualization
tensor_3d_np = tensor_3d_tf.numpy()

# Visualize the slices of the 3D tensor
plot_tensor_slices(tensor_3d_np, axis=0)  # Slicing along axis 0
```

```python
def plot_3d_scatter(tensor):
    """ Function to create a 3D scatter plot of a 3D tensor's values.
    tensor (np.ndarray): A 3D tensor to visualize in scatter plot form.
    """
    assert tensor.ndim == 3, "The input tensor must be 3-dimensional."

    # Get the indices and values of the tensor
    x, y, z = np.indices(tensor.shape)
    values = tensor.flatten()

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Normalize values for coloring
    colors = plt.cm.viridis(values / values.max())

    # Create a 3D scatter plot
    ax.scatter(x.flatten(), y.flatten(), z.flatten(), c=colors, s=50)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.title('3D Scatter Plot of Tensor Values')
    plt.show()

# Visualize the 3D tensor using a scatter plot
plot_3d_scatter(tensor_3d_np)
```
